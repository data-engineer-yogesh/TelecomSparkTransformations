### Unity Catalog Design 
name -> telecomsparktransformations_catalog

**Disclaimer** : In our free Databricks environment, we follow the Unity Catalog structure by creating a catalog and schema, then a volume, where we upload the necessary files for our projects

**First we list the existing catalogs.**

```python
%sql 

-- 
ShOW CATALOGS;
```

**Let's create the telecomsparktransformations_catalog**

```python
%sql
CREATE CATALOG telecomsparktransformations_catalog;

COMMENT ON CATALOG telecomsparktransformations_catalog IS'we don t have extrnal location';
```

**Here we check the current catalog is in use.** 

```python
%sql
 

 SELECT current_catalog();
```

**We need to select our catalog telecomsparktransformations_catalog for use.**

```python
%sql 
-- 

USE CATALOG telecomsparktransformations_catalog;
     
  SELECT current_catalog();
```

**Get more info about the telecomsparktransformations_catalog.**

```python
%sql 

DESCRIBE CATALOG EXTENDED telecomsparktransformations_catalog;
```

**We create all schemas under the catalog telecomsparktransformations_catalog.**

```python
%sql

CREATE SCHEMA IF NOT EXISTS Bronze;

CREATE SCHEMA IF NOT EXISTS Silver;

CREATE SCHEMA IF NOT EXISTS Gold;

CREATE SCHEMA IF NOT EXISTS landing;
```

**List the schemas we created.**

```python
%sql 

SHOW SCHEMAS;
```

```python
%sql 
USE SCHEMA landing;
     
CREATE VOLUME If NOT EXISTS telecomsparktransformations_catalog.landing.operational_data;
```
